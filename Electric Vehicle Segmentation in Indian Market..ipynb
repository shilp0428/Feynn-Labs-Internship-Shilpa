```python
df.rename(columns = {'Range':'Range(Km/Full Charge)'},inplace = True)
df
```

```python
df['Range(Km/Full Charge)'] = df['Range(Km/Full Charge)'].apply(lambda x: str(x).split()[0])
```

```python
df.replace('na',np.nan,inplace = True)
df.info()
```

```python
df = df.iloc[:,:-3]
df['Range(Km/Full Charge)'] = df['Range(Km/Full Charge)'].astype(float)
df.info()
```

```python
df['Range(Km/Full Charge)'].replace(np.nan,round(df['Range(Km/Full Charge)'].mean()),inplace = True)
df.info()
```

```python
df['PriceRange'].replace(" ","",inplace = True)
df
```

```python
import numpy as np

df['Top Price(L)'] = 0
for i in range(len(df)):
    x = df.loc[i, 'PriceRange']
    if isinstance(x, str):
        x = x.split()
        if (len(x) > 3) and (x[-1] == 'L'):
            df.loc[i, 'Base Price(L)'] = float(x[1])
    else:
        df.loc[i, 'Base Price(L)'] = np.nan
```

```python
df
```

```python
import numpy as np

df['Capacity'] = df['Capacity'].apply(lambda x: int(str(x).split()[0]) if isinstance(x, str) else np.nan)
df.drop(['PriceRange', 'Transmission', 'VehicleType'], axis=1, inplace=True)
df.head()
```

## EDA

```python
import re

# Loop over the column names
for col in num_cols:
    # Clean the column by extracting the numeric part
    df[col] = df[col].apply(lambda x: float(re.findall(r'\d+', str(x))[0]) if isinstance(x, str) else x)
    
    # Plot the distribution after cleaning the data
    sns.distplot(df[col])
    
# Rest of your code for plotting
plt.tight_layout()
plt.show()
```

```python

plt.figure(figsize = (12,8))
for i in range(len(num_cols)):
  plt.subplot(len(num_cols)//2,2,i+1)
  sns.distplot(df[num_cols[i]])
plt.tight_layout()
plt.show()
     
```

```python

for i in num_cols:
  plt.figure(figsize = (15,5))
  sns.barplot(data = df,x = 'Style',y = i)
  plt.title('Car style and {}'.format(i))
  plt.xticks(rotation = 45)
  plt.show()
```
